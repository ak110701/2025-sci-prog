# 05 — Završni izvještaj: Predviđanje posjećenosti skijališta

## Cilj projekta
Cilj ovog projekta je izraditi **proxy varijablu posjećenosti skijališta**
(*low / medium / high*) na temelju:

- snježnih uvjeta
- vremenskih značajki (mjesec, sezona)
- karakteristika skijališta (broj staza, liftova, cijena, lokacija)

Na temelju tih podataka trenirani su modeli strojnog učenja za predviđanje
razine posjećenosti.

```python
import warnings
warnings.filterwarnings("ignore")

from pathlib import Path
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

PROJECT_ROOT = Path.cwd().parent
DATA_PATH = PROJECT_ROOT / "data" / "processed" / "features.csv"

REPORTS_DIR = PROJECT_ROOT / "reports"
FIG_DIR = REPORTS_DIR / "figures"
REPORTS_DIR.mkdir(exist_ok=True)
FIG_DIR.mkdir(parents=True, exist_ok=True)

print("PROJECT_ROOT:", PROJECT_ROOT)
print("DATA_PATH exists:", DATA_PATH.exists(), "|", DATA_PATH)
```

```python
df = pd.read_csv(DATA_PATH)

print("Rows:", len(df), "Cols:", df.shape[1])
display(df.head(3))

assert len(df) > 0, "features.csv is empty"
assert "attendance" in df.columns, "Target 'attendance' missing"
print("Target distribution:\n", df["attendance"].value_counts())
```

## Eksplorativni uvidi (kratko)

Analizirani su sljedeći aspekti:
- raspodjela razina posjećenosti
- raspodjela intenziteta snježnih uvjeta
- sezonalnost (po mjesecima)
- razlike između država i pojedinih skijališta

```python
# Attendance distribution
ax = df["attendance"].value_counts().plot(kind="bar")
ax.set_title("Attendance (target) distribution")
plt.tight_layout()
plt.savefig(FIG_DIR / "target_distribution.png", dpi=200)
plt.show()

# Snow distribution (use whatever exists)
snow_candidates = [c for c in ["snow", "snow_mean", "snow_7d_mean", "snow_7d_sum"] if c in df.columns]
print("Snow-like cols found:", snow_candidates)

if snow_candidates:
    col = snow_candidates[0]
    ax = df[col].dropna().plot(kind="hist", bins=30)
    ax.set_title(f"Distribution of {col}")
    plt.tight_layout()
    plt.savefig(FIG_DIR / f"{col}_distribution.png", dpi=200)
    plt.show()
```

```python
# robust month column
month_col = None
if "month_num" in df.columns:
    month_col = "month_num"
elif "month" in df.columns:
    # could be string date -> try parse
    m = pd.to_datetime(df["month"], errors="coerce")
    if m.notna().any():
        df["_month_num_tmp"] = m.dt.month
        month_col = "_month_num_tmp"

if month_col:
    tmp = (
        df.groupby([month_col, "attendance"])
        .size()
        .reset_index(name="count")
        .pivot(index=month_col, columns="attendance", values="count")
        .fillna(0)
    )
    ax = tmp.plot(kind="bar", stacked=True)
    ax.set_title("Attendance by month (stacked counts)")
    plt.tight_layout()
    plt.savefig(FIG_DIR / "attendance_by_month.png", dpi=200)
    plt.show()
else:
    print("No usable month column found.")
```

```python
from sklearn.model_selection import train_test_split
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.impute import SimpleImputer

from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier, HistGradientBoostingClassifier
from sklearn.svm import LinearSVC
TARGET = "attendance"

# drop leakage / non-features
drop_cols = [TARGET]
# if month is a string date, we usually keep month_num instead
if "month" in df.columns and "month_num" in df.columns:
    drop_cols.append("month")

X = df.drop(columns=[c for c in drop_cols if c in df.columns]).copy()
y = df[TARGET].astype(str).copy()

print("X shape:", X.shape, "y shape:", y.shape)
print("y distribution:\n", y.value_counts())

X_train, X_test, y_train, y_test = train_test_split(
    X, y,
    test_size=0.30,
    random_state=42,
    stratify=y
)

# detect numeric/categorical
num_cols = X_train.select_dtypes(include=["number"]).columns.tolist()
cat_cols = [c for c in X_train.columns if c not in num_cols]

print("Numeric cols:", len(num_cols))
print("Categorical cols:", len(cat_cols))

numeric_pipe = Pipeline(steps=[
    ("imputer", SimpleImputer(strategy="median")),
    ("scaler", StandardScaler())
])

categorical_pipe = Pipeline(steps=[
    ("imputer", SimpleImputer(strategy="most_frequent")),
    ("ohe", OneHotEncoder(handle_unknown="ignore"))
])

preprocess = ColumnTransformer(
    transformers=[
        ("num", numeric_pipe, num_cols),
        ("cat", categorical_pipe, cat_cols),
    ],
    remainder="drop",
    sparse_threshold=0.0  # force dense to avoid HGB sparse error
)
```

```python
from sklearn.metrics import classification_report, confusion_matrix, ConfusionMatrixDisplay

def evaluate(model, X_test, y_test, title="Model"):
    y_pred = model.predict(X_test)
    print(f"\n=== {title} ===")
    print(classification_report(y_test, y_pred))
    cm = confusion_matrix(y_test, y_pred, labels=np.unique(y_test))
    disp = ConfusionMatrixDisplay(cm, display_labels=np.unique(y_test))
    disp.plot()
    plt.title(title)
    plt.tight_layout()
    plt.savefig(FIG_DIR / f"cm_{title.replace(' ','_').lower()}.png", dpi=200)
    plt.show()
    return y_pred
```

```python
from sklearn.dummy import DummyClassifier

baseline = Pipeline(steps=[
    ("prep", preprocess),
    ("model", DummyClassifier(strategy="most_frequent"))
])

baseline.fit(X_train, y_train)
_ = evaluate(baseline, X_test, y_test, title="Baseline (Most Frequent)")
```

```python
logreg = Pipeline(steps=[
    ("prep", preprocess),
    ("model", LogisticRegression(max_iter=2000))
])

logreg.fit(X_train, y_train)
_ = evaluate(logreg, X_test, y_test, title="Logistic Regression")
```

```python
rf = Pipeline(steps=[
    ("prep", preprocess),
    ("model", RandomForestClassifier(
        n_estimators=400,
        random_state=42,
        class_weight="balanced",
        n_jobs=-1
    ))
])

rf.fit(X_train, y_train)
_ = evaluate(rf, X_test, y_test, title="Random Forest")
```

```python
from sklearn.svm import LinearSVC

svc = Pipeline(steps=[
    ("prep", preprocess),
    ("model", LinearSVC())
])

svc.fit(X_train, y_train)
_ = evaluate(svc, X_test, y_test, title="LinearSVC")
```

## Uvid u značajke (feature insights)

Kod modela koji koriste one-hot kodirane varijable, klasična analiza važnosti
značajki najjednostavnije se provodi pomoću modela temeljenih na stablima.

Umjesto složene globalne analize važnosti, korištena je:
- **permutation importance** metoda na najboljem modelu (HGB)

Ova metoda daje interpretabilan poredak značajki i pokazuje koji stupci
najviše utječu na točnost predviđanja.

```python
from sklearn.inspection import permutation_importance
import pandas as pd
import numpy as np

best_model = rf  # tvoj pipeline (prep + model)

result = permutation_importance(
    best_model,
    X_test,
    y_test,
    n_repeats=10,
    random_state=42,
    n_jobs=-1,
    scoring="accuracy",
)

prep = best_model.named_steps["prep"]

# 1) stvarni broj featurea iz permutation_importance
n_imp = result.importances_mean.shape[0]

# 2) probaj dohvatit imena
try:
    feature_names = prep.get_feature_names_out()
    feature_names = np.asarray(feature_names)
except Exception:
    feature_names = np.asarray([f"feature_{i}" for i in range(n_imp)])

# 3) ako se duljine ne poklapaju, poravnaj
if len(feature_names) != n_imp:
    print(f"⚠️ Mismatch: names={len(feature_names)} vs importances={n_imp}. Poravnavam na min().")
    m = min(len(feature_names), n_imp)
    feature_names = feature_names[:m]
    import_mean = result.importances_mean[:m]
    import_std  = result.importances_std[:m]
else:
    import_mean = result.importances_mean
    import_std  = result.importances_std

imp = pd.DataFrame({
    "feature": feature_names,
    "importance_mean": import_mean,
    "importance_std": import_std,
}).sort_values("importance_mean", ascending=False)

display(imp.head(30))
```
