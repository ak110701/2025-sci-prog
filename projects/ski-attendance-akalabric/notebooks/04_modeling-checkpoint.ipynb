# 04 — Modeling: Attendance Prediction

Cilj:
- trenirati baseline i ML modele za predviđanje razine posjećenosti (attendance)
- evaluirati modele pomoću klasifikacijskih metrika
- analizirati feature importance i ograničenja modela

Target varijabla:
- attendance ∈ {low, medium, high}

Dataset:
- features.csv (generiran u 03_feature_engineering)

```python
import numpy as np
import pandas as pd

from pathlib import Path
import matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, confusion_matrix, ConfusionMatrixDisplay

from sklearn.dummy import DummyClassifier

from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.impute import SimpleImputer

from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier, HistGradientBoostingClassifier
from sklearn.svm import LinearSVC
```

```python
def find_project_root(start: Path | None = None) -> Path:
    """Find project root by walking upwards until we find 'data' folder."""
    start = start or Path.cwd()
    cur = start.resolve()
    for _ in range(10):
        if (cur / "data").exists():
            return cur
        cur = cur.parent
    return start.resolve()

ROOT = find_project_root()
DATA_PROCESSED = ROOT / "data" / "processed"
FEATURES_PATH = DATA_PROCESSED / "features.csv"

print("ROOT:", ROOT)
print("FEATURES_PATH exists:", FEATURES_PATH.exists())
print("FEATURES_PATH:", FEATURES_PATH)
```

```python
df = pd.read_csv(FEATURES_PATH)

print("Loaded rows:", len(df))
print("Cols:", df.shape[1])
display(df.head(3))
```

```python
TARGET = "attendance"

if TARGET not in df.columns:
    raise ValueError(f"Target '{TARGET}' ne postoji u features.csv. Imam stupce: {df.columns.tolist()}")

print("Target distribution:")
print(df[TARGET].astype(str).value_counts(dropna=False))
```

```python
df = df.copy()

# target mora biti prisutan
df[TARGET] = df[TARGET].astype(str)
df = df[df[TARGET].notna()].copy()

# ako postoji month, pretvori i napravi year/month_num pa originalni month izbaci iz feature seta
if "month" in df.columns:
    df["month"] = pd.to_datetime(df["month"], errors="coerce")
    # ako ima NaT, nije smak svijeta — ali ćemo ga kasnije imputat kroz categorical ili drop
    df["month_year"] = df["month"].dt.year
    df["month_num"] = df["month"].dt.month
    # original month je datetime -> ne smije direktno u sklearn modele
    # ostavi ga za pregled, ali ga kasnije dropamo iz X
    print("month parsed. NaT count:", df["month"].isna().sum())

display(df[[c for c in ["month","month_year","month_num",TARGET] if c in df.columns]].head(5))
```

```python
DROP_ALWAYS = [TARGET]
if "month" in df.columns:
    DROP_ALWAYS.append("month")

# ove meta kolone možeš ostaviti, ali često su OK kao category;
# ja ih NE dropam automatski osim ako ti želiš:
OPTIONAL_DROP = []  # npr. ["id"] ako misliš da je leak

drop_cols = [c for c in DROP_ALWAYS + OPTIONAL_DROP if c in df.columns]

X = df.drop(columns=drop_cols)
y = df[TARGET].astype(str)

print("X shape:", X.shape)
print("y shape:", y.shape)
print("y classes:", sorted(y.unique()))
```

```python
# safety: ako neka klasa ima premalo uzoraka, stratify može puknuti
vc = y.value_counts()
print("Class counts:\n", vc)

if (vc < 2).any():
    print("⚠️ Neka klasa ima <2 uzorka -> gasim stratify (inače split puca).")
    strat = None
else:
    strat = y

X_train, X_test, y_train, y_test = train_test_split(
    X, y,
    test_size=0.30,
    random_state=42,
    stratify=strat
)

print("Train:", X_train.shape, "Test:", X_test.shape)
```

```python
num_cols = X.select_dtypes(include=[np.number]).columns.tolist()
cat_cols = [c for c in X.columns if c not in num_cols]

print("Numeric cols:", len(num_cols))
print("Categorical cols:", len(cat_cols))
print("Example cat cols:", cat_cols[:10])
```

```python
numeric_pipe = Pipeline(steps=[
    ("imputer", SimpleImputer(strategy="median")),
    ("scaler", StandardScaler())
])

categorical_pipe = Pipeline(steps=[
    ("imputer", SimpleImputer(strategy="most_frequent")),
    ("onehot", OneHotEncoder(handle_unknown="ignore"))
])

preprocess = ColumnTransformer(
    transformers=[
        ("num", numeric_pipe, num_cols),
        ("cat", categorical_pipe, cat_cols),
    ],
    remainder="drop"
)
```

```python
def evaluate(model, X_test, y_test, title="MODEL"):
    pred = model.predict(X_test)
    print(f"\n=== {title} ===")
    print(classification_report(y_test, pred))
    
    labels = sorted(np.unique(y_test))
    cm = confusion_matrix(y_test, pred, labels=labels)
    disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=labels)
    disp.plot()
    plt.title(title)
    plt.show()
    return pred
```

```python
baseline = Pipeline(steps=[
    ("prep", preprocess),
    ("model", DummyClassifier(strategy="most_frequent"))
])

baseline.fit(X_train, y_train)
_ = evaluate(baseline, X_test, y_test, title="Baseline: Most Frequent")
```

```python
logreg = Pipeline(steps=[
    ("prep", preprocess),
    ("model", LogisticRegression(max_iter=2000))
])

logreg.fit(X_train, y_train)
_ = evaluate(logreg, X_test, y_test, title="Logistic Regression")
```

```python
rf = Pipeline(steps=[
    ("prep", preprocess),
    ("model", RandomForestClassifier(
        n_estimators=400,
        random_state=42,
        class_weight="balanced",
        n_jobs=-1
    ))
])

rf.fit(X_train, y_train)
_ = evaluate(rf, X_test, y_test, title="Random Forest")
```

```python
# ==========================
# Model 3: Linear SVM (sparse-friendly)
# ==========================

svm = Pipeline(steps=[
    ("prep", preprocess),
    ("model", LinearSVC(random_state=42))
])

svm.fit(X_train, y_train)
_ = evaluate(svm, X_test, y_test, title="LinearSVC (SVM)")
```

```python
# Dohvati feature name-ove nakon one-hot
ohe = rf.named_steps["prep"].named_transformers_["cat"].named_steps["onehot"]
cat_feature_names = ohe.get_feature_names_out(cat_cols).tolist() if len(cat_cols) else []
all_feature_names = num_cols + cat_feature_names

importances = rf.named_steps["model"].feature_importances_
fi = pd.DataFrame({"feature": all_feature_names, "importance": importances})
fi = fi.sort_values("importance", ascending=False)

display(fi.head(20))

# plot top 20
top = fi.head(20).iloc[::-1]
plt.figure()
plt.barh(top["feature"], top["importance"])
plt.title("RandomForest Feature Importance (Top 20)")
plt.xlabel("Importance")
plt.show()
```
