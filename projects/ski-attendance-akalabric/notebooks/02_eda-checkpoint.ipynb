# 02 — Exploratory Data Analysis (EDA)
Ovaj notebook istražuje integrirani skup extracted_snow.csv koji sadrži podatke po skijalištu i mjesecu (resort × month). Cilj EDA-e je:

- razumjeti raspodjele (snijeg, visina, cijena, liftovi, staze)

- otkriti sezonalnost (month → snow)

- provjeriti missing vrijednosti i kvalitetu spajanja

- dobiti input za feature engineering i modeliranje

```python
import sys
print(sys.executable)
```

```python
import warnings
warnings.filterwarnings("ignore")

import numpy as np
import pandas as pd

from pathlib import Path

import matplotlib.pyplot as plt
```

```python
from pathlib import Path

def find_project_root(start: Path) -> Path:
    cur = start.resolve()
    for _ in range(10):
        if (cur / "README.md").exists() and (cur / "notebooks").exists():
            return cur
        cur = cur.parent
    return start.resolve()

CWD = Path.cwd()
PROJECT_ROOT = find_project_root(CWD)
DATA_DIR = PROJECT_ROOT / "data"
RAW_DIR = DATA_DIR / "raw"
PROCESSED_DIR = DATA_DIR / "processed"

DATA_PATH = PROCESSED_DIR / "extracted_snow.csv"

print("CWD:", CWD)
print("PROJECT_ROOT:", PROJECT_ROOT)
print("DATA_PATH exists:", DATA_PATH.exists())
print("DATA_PATH:", DATA_PATH.resolve())
```

```python
# encoding fallback (jer si već imala unicode issues)
encodings_to_try = ["utf-8", "utf-8-sig", "cp1252", "latin1"]

df = None
last_err = None
for enc in encodings_to_try:
    try:
        df = pd.read_csv(DATA_PATH, encoding=enc)
        print("Loaded with encoding:", enc)
        break
    except Exception as e:
        last_err = e

if df is None:
    raise last_err

print("Rows:", len(df))
print("Columns:", df.shape[1])
display(df.head(3))
```

```python
df = df.copy()

# normalize columns (ako ima čudnih razmaka)
df.columns = (
    df.columns.astype(str)
    .str.strip()
    .str.replace(" ", "_")
    .str.lower()
)

# očekujemo month kao int (1-12)
if "month" in df.columns:
    df["month"] = pd.to_datetime(df["month"], errors="coerce")

# snow kao numeric
if "snow" in df.columns:
    df["snow"] = pd.to_numeric(df["snow"], errors="coerce")

# id kao int ako postoji
if "id" in df.columns:
    df["id"] = pd.to_numeric(df["id"], errors="coerce").astype("Int64")

# najčešće numeric kolone
numeric_candidates = [
    "price","highest_point","lowest_point",
    "beginner_slopes","intermediate_slopes","difficult_slopes","total_slopes",
    "surface_lifts","chair_lifts","gondola_lifts","total_lifts",
    "lift_capacity","latitude","longitude"
]
for c in numeric_candidates:
    if c in df.columns:
        df[c] = pd.to_numeric(df[c], errors="coerce")

print("Dtypes snapshot:")
display(df.dtypes.head(25))
```

## Struktura i ključne dimenzije

U ovom datasetu očekujemo da je osnovna granularnost (resort, month).
Provjeravamo:

- koliko imamo skijališta

- koliko mjeseci

- duplikate (resort-month)

- raspon po kontinentima i državama

```python
n_rows = len(df)
n_resorts = df["id"].nunique() if "id" in df.columns else df["resort"].nunique()
n_months = df["month"].nunique() if "month" in df.columns else None

print("Rows:", n_rows)
print("Unique resorts:", n_resorts)
print("Unique months:", n_months)

# duplicate check on (id, month)
if "id" in df.columns and "month" in df.columns:
    dup = df.duplicated(subset=["id","month"]).sum()
    print("Duplicates (id,month):", dup)
    if dup > 0:
        display(df[df.duplicated(subset=["id","month"], keep=False)].head(10))
```

```python
if "continent" in df.columns:
    print("\nTop continents by rows:")
    display(df["continent"].value_counts().head(10))

if "country" in df.columns:
    print("\nTop countries by resorts:")
    if "id" in df.columns:
        display(df.groupby("country")["id"].nunique().sort_values(ascending=False).head(15))
    else:
        display(df["country"].value_counts().head(15))
```

## Missing values / kvaliteta spajanja

Snijeg može biti missing zbog grida, rubnih lokacija ili fallback limita.
Za modeliranje ćemo kasnije:

- imati flag snow_missing

- fill s 0 ili median po kontinentu/mjesecu (ovisno o scenariju

```python
miss = df.isna().mean().sort_values(ascending=False)
miss = (miss * 100).round(2)

print("Top missing columns (%):")
display(miss.head(20))

if "snow" in df.columns:
    print("\nMissing snow %:", float(df["snow"].isna().mean() * 100))
```

```python
if "snow" in df.columns and "month" in df.columns:
    miss_by_month = df.groupby("month")["snow"].apply(lambda s: s.isna().mean()).reset_index(name="missing_ratio")
    plt.figure()
    plt.plot(miss_by_month["month"], miss_by_month["missing_ratio"])
    plt.title("Missing snow ratio by month")
    plt.xlabel("Month")
    plt.ylabel("Missing ratio")
    plt.show()

if "snow" in df.columns and "continent" in df.columns:
    miss_by_cont = df.groupby("continent")["snow"].apply(lambda s: s.isna().mean()).sort_values(ascending=False)
    plt.figure()
    plt.bar(miss_by_cont.index.astype(str), miss_by_cont.values)
    plt.title("Missing snow ratio by continent")
    plt.xticks(rotation=45, ha="right")
    plt.ylabel("Missing ratio")
    plt.show()
```

## Distribucije: snow, visina, cijena, kapacitet

Cilj: vidjeti ima li outliera i kakva je skala.

```python
cols_to_describe = [c for c in ["snow","highest_point","lowest_point","price","lift_capacity","total_slopes","total_lifts"] if c in df.columns]
display(df[cols_to_describe].describe(include="all"))
```

```python
def hist(col, bins=40):
    if col not in df.columns:
        return
    series = df[col].dropna()
    plt.figure()
    plt.hist(series, bins=bins)
    plt.title(f"Histogram: {col}")
    plt.xlabel(col)
    plt.ylabel("count")
    plt.show()

for c in ["snow","highest_point","lowest_point","price","lift_capacity","total_slopes","total_lifts"]:
    hist(c)
```

## Sezonalnost: snijeg kroz mjesece

Očekujemo da će winter months imati veći snijeg (ovisno o hemisferi i kontinentu).
Budući da je dataset globalan, vidjet ćemo i “razmazani” signal, ali Europa bi trebala imati jasan pattern.

```python
if "snow" in df.columns and "month" in df.columns:
    tmp = df.copy()
    tmp["snow_filled"] = tmp["snow"].fillna(0)

    mean_by_month = tmp.groupby("month")["snow_filled"].mean().reset_index(name="snow_mean")
    plt.figure()
    plt.plot(mean_by_month["month"], mean_by_month["snow_mean"])
    plt.title("Average snow by month (global, NaN->0)")
    plt.xlabel("Month")
    plt.ylabel("Average snow")
    plt.show()

    if "continent" in tmp.columns:
        eur = tmp[tmp["continent"] == "Europe"].copy()
        mean_eur = eur.groupby("month")["snow_filled"].mean().reset_index(name="snow_mean")
        plt.figure()
        plt.plot(mean_eur["month"], mean_eur["snow_mean"])
        plt.title("Average snow by month (Europe, NaN->0)")
        plt.xlabel("Month")
        plt.ylabel("Average snow")
        plt.show()
```

## “Resort profile”: visina ↔ snijeg

Hipoteza: veća nadmorska visina → više snijega (prosječno).
Radimo jednostavnu korelaciju i scatter.

```python
if all(c in df.columns for c in ["highest_point","snow"]):
    tmp = df[["highest_point","snow"]].dropna().copy()
    tmp = tmp.sample(min(4000, len(tmp)), random_state=42)  # da ne crta milijun točaka

    plt.figure()
    plt.scatter(tmp["highest_point"], tmp["snow"], s=10)
    plt.title("Scatter: highest_point vs snow")
    plt.xlabel("highest_point")
    plt.ylabel("snow")
    plt.show()

    corr = df[["highest_point","snow"]].corr(numeric_only=True).iloc[0,1]
    print("Correlation(highest_point, snow):", corr)
```

## Korelacijska matrica (numeric)
Bez seaborn: koristimo matplotlib imshow.

```python
numeric_df = df.select_dtypes(include=[np.number]).copy()

# izbacimo ID-eve da ne kvare
for c in ["id","month"]:
    if c in numeric_df.columns:
        pass

corr = numeric_df.corr()

plt.figure(figsize=(10, 8))
plt.imshow(corr.values)
plt.title("Correlation matrix (numeric)")
plt.xticks(range(len(corr.columns)), corr.columns, rotation=90)
plt.yticks(range(len(corr.columns)), corr.columns)
plt.colorbar()
plt.tight_layout()
plt.show()
```

## “Najsnježnija” skijališta (po prosjeku)

Ovo je korisno za sanity-check i priču u reportu.

```python
if all(c in df.columns for c in ["resort","snow"]):
    tmp = df.copy()
    tmp["snow_filled"] = tmp["snow"].fillna(0)

    top = (
        tmp.groupby(["resort","country"], dropna=False)["snow_filled"]
        .mean()
        .sort_values(ascending=False)
        .head(15)
        .reset_index()
    )
    display(top)
```
