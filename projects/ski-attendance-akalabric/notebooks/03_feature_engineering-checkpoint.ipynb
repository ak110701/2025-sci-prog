# 03 — Feature Engineering (All resorts × month)
U ovom notebooku:

- čistimo i standardiziramo podatke iz extracted_snow.csv

- radimo time featuree (month cyclical)

- radimo snow featuree (missing flag + imputacija + rolling po resortu)

- radimo resort profile featuree (visina, staze, liftovi, cijena…)

- radimo proxy “demand/attendance” score i klasificiramo u attendance (low/medium/high)

- spremamo rezultat kao data/processed/features.csv

```python
import warnings
warnings.filterwarnings("ignore")

import numpy as np
import pandas as pd

from pathlib import Path
import matplotlib.pyplot as plt
```

```python
def find_project_root(start: Path) -> Path:
    cur = start.resolve()
    for _ in range(10):
        if (cur / "README.md").exists() and (cur / "notebooks").exists():
            return cur
        cur = cur.parent
    return start.resolve()

CWD = Path.cwd()
PROJECT_ROOT = find_project_root(CWD)

DATA_DIR = PROJECT_ROOT / "data"
RAW_DIR = DATA_DIR / "raw"
PROCESSED_DIR = DATA_DIR / "processed"

IN_PATH = PROCESSED_DIR / "extracted_snow.csv"
OUT_PATH = PROCESSED_DIR / "features.csv"

print("CWD:", CWD)
print("PROJECT_ROOT:", PROJECT_ROOT)
print("IN exists:", IN_PATH.exists(), IN_PATH.resolve())
print("OUT will be:", OUT_PATH.resolve())
```

```python
encodings_to_try = ["utf-8", "utf-8-sig", "cp1252", "latin1"]
df = None
last_err = None

for enc in encodings_to_try:
    try:
        df = pd.read_csv(IN_PATH, encoding=enc)
        print("Loaded with encoding:", enc)
        break
    except Exception as e:
        last_err = e

if df is None:
    raise last_err

# normalize columns
df.columns = (
    df.columns.astype(str)
    .str.strip()
    .str.replace(" ", "_")
    .str.lower()
)

print("Rows:", len(df), "Cols:", df.shape[1])
display(df.head(3))
```

### Minimalna validacija + tipovi

Očekujemo stupce:

- id, resort, country, continent

- month

- snow (može biti missing)

i opcionalno: price, highest_point, total_slopes, total_lifts, itd.

```python
# =========================
# FIX: month je datum (robustno parsiranje)
# =========================

# required columns
required = ["month", "snow"]
missing_req = [c for c in required if c not in df.columns]
if missing_req:
    raise ValueError(f"Fale obavezni stupci: {missing_req}. Imam: {df.columns.tolist()}")

# --- clean month strings (BOM/razmaci)
m = df["month"].astype(str).str.strip()
m = m.str.replace("\ufeff", "", regex=False)

# --- attempt 1: normal datetime parse
dt = pd.to_datetime(m, errors="coerce")

# --- attempt 2: ako je puno NaT (npr. "2022-01" ili "01.2022")
if dt.isna().mean() > 0.5:
    m2 = m.copy()
    # "YYYY-MM" -> "YYYY-MM-01"
    m2 = m2.str.replace(r"^(\d{4})-(\d{2})$", r"\1-\2-01", regex=True)
    # "MM.YYYY" -> "YYYY-MM-01"
    m2 = m2.str.replace(r"^(\d{2})\.(\d{4})$", r"\2-\1-01", regex=True)
    dt = pd.to_datetime(m2, errors="coerce")

df["month"] = dt  # month kao datetime

# numeric snow
df["snow"] = pd.to_numeric(df["snow"], errors="coerce")

# optional id
if "id" in df.columns:
    df["id"] = pd.to_numeric(df["id"], errors="coerce").astype("Int64")

# drop invalid months
before = len(df)
df = df.dropna(subset=["month"]).copy()
after = len(df)
print("Dropped invalid month rows:", before - after)

# extract time features
df["month_num"] = df["month"].dt.month.astype(int)
df["year"] = df["month"].dt.year.astype(int)

print("Rows after month parsing:", len(df))
print("Month range:", df["month"].min(), "->", df["month"].max())

display(df[["month", "month_num", "year", "snow"]].head())
```

### Ključ (resort, month) i duplikati

Ako ima duplikata, agregiramo (mean za snow, first za meta).

```python
key_cols = []
if "id" in df.columns:
    key_cols = ["id", "month"]
else:
    key_cols = ["resort", "month"]

dup_count = df.duplicated(subset=key_cols).sum()
print("Duplicates on key:", dup_count)

if dup_count > 0:
    # define aggregation
    agg = {}
    for col in df.columns:
        if col in key_cols:
            continue
        if pd.api.types.is_numeric_dtype(df[col]):
            agg[col] = "mean"
        else:
            agg[col] = "first"
    df = df.groupby(key_cols, as_index=False).agg(agg)
    print("After aggregation rows:", len(df))
```

### Snow featurei

- snow_missing (0/1)

- imputacija snijega (median po kontinentu + month, fallback global median)

- rolling featurei po resortu kroz mjesece:

    - snow_lag_1, snow_lag_2

    - snow_3m_mean, snow_3m_std

    - snow_6m_mean

Pošto je mjesec (1..12) cikličan, “rolling” radimo nakon sortiranja po month unutar resorta.

```python
df = df.copy()

df["snow_missing"] = df["snow"].isna().astype(int)

# helper median tables
global_median = float(df["snow"].median(skipna=True)) if df["snow"].notna().any() else 0.0

if "continent" in df.columns:
    med = (
        df.groupby(["continent","month"])["snow"]
        .median()
        .reset_index(name="snow_med_cm")
    )
    df = df.merge(med, on=["continent","month"], how="left")
    df["snow_imputed"] = df["snow"].fillna(df["snow_med_cm"])
    df["snow_imputed"] = df["snow_imputed"].fillna(global_median)
    df = df.drop(columns=["snow_med_cm"])
else:
    # fallback: month median only
    med_m = df.groupby("month")["snow"].median().reset_index(name="snow_med_m")
    df = df.merge(med_m, on="month", how="left")
    df["snow_imputed"] = df["snow"].fillna(df["snow_med_m"]).fillna(global_median)
    df = df.drop(columns=["snow_med_m"])

print("Missing snow before:", df["snow"].isna().mean())
print("Missing snow_imputed:", df["snow_imputed"].isna().mean())
print("Global median used:", global_median)
```

```python
# choose group id
gid = "id" if "id" in df.columns else "resort"

df = df.sort_values([gid, "month"]).copy()

# lags
df["snow_lag_1"] = df.groupby(gid)["snow_imputed"].shift(1)
df["snow_lag_2"] = df.groupby(gid)["snow_imputed"].shift(2)

# rolling windows (min_periods=1 so we don't create NaN-only)
df["snow_3m_mean"] = df.groupby(gid)["snow_imputed"].transform(lambda s: s.rolling(3, min_periods=1).mean())
df["snow_3m_std"]  = df.groupby(gid)["snow_imputed"].transform(lambda s: s.rolling(3, min_periods=1).std())
df["snow_6m_mean"] = df.groupby(gid)["snow_imputed"].transform(lambda s: s.rolling(6, min_periods=1).mean())

# std can be NaN when window size =1; fill with 0 (no variability yet)
df["snow_3m_std"] = df["snow_3m_std"].fillna(0.0)

display(df[[gid, "month", "snow", "snow_imputed", "snow_lag_1", "snow_3m_mean"]].head(10))
```

### Time featurei (cyclical)

Month je cikličan → koristimo sin/cos.

```python
m = df["month"].astype(int)

df["month_sin"] = np.sin(2 * np.pi * m / 12.0)
df["month_cos"] = np.cos(2 * np.pi * m / 12.0)

# simple season flags (Europe-friendly; ali ostaje kao feature)
df["is_winter_month"] = df["month_num"].isin([12,1,2,3]).astype(int)
df["is_shoulder_month"] = df["month_num"].isin([11,4]).astype(int)

display(df[["month","month_sin","month_cos","is_winter_month"]].drop_duplicates().sort_values("month").head(12))
```

### Resort profile featurei

Radimo stabilne featuree iz:

- visine (range)

- staze (ukupno i udjeli)

- liftovi (ukupno i udjeli)

- cijena (log transform)

```python
df = df.copy()

# altitude features
if all(c in df.columns for c in ["highest_point","lowest_point"]):
    df["altitude_range"] = df["highest_point"] - df["lowest_point"]
else:
    df["altitude_range"] = np.nan

# slopes: totals + shares
slope_cols = ["beginner_slopes","intermediate_slopes","difficult_slopes","total_slopes"]
for c in slope_cols:
    if c in df.columns:
        df[c] = pd.to_numeric(df[c], errors="coerce")

if all(c in df.columns for c in ["beginner_slopes","intermediate_slopes","difficult_slopes"]):
    df["slopes_sum_calc"] = df["beginner_slopes"].fillna(0) + df["intermediate_slopes"].fillna(0) + df["difficult_slopes"].fillna(0)
    # if total_slopes missing, use calc
    if "total_slopes" not in df.columns:
        df["total_slopes"] = df["slopes_sum_calc"]
    df["total_slopes"] = df["total_slopes"].fillna(df["slopes_sum_calc"])
else:
    df["slopes_sum_calc"] = np.nan

if "total_slopes" in df.columns:
    denom = df["total_slopes"].replace(0, np.nan)
    if "beginner_slopes" in df.columns:
        df["share_beginner"] = (df["beginner_slopes"] / denom).fillna(0)
    else:
        df["share_beginner"] = np.nan
    if "intermediate_slopes" in df.columns:
        df["share_intermediate"] = (df["intermediate_slopes"] / denom).fillna(0)
    else:
        df["share_intermediate"] = np.nan
    if "difficult_slopes" in df.columns:
        df["share_difficult"] = (df["difficult_slopes"] / denom).fillna(0)
    else:
        df["share_difficult"] = np.nan

# lifts: totals + shares
lift_cols = ["surface_lifts","chair_lifts","gondola_lifts","total_lifts","lift_capacity"]
for c in lift_cols:
    if c in df.columns:
        df[c] = pd.to_numeric(df[c], errors="coerce")

if all(c in df.columns for c in ["surface_lifts","chair_lifts","gondola_lifts"]):
    df["lifts_sum_calc"] = df["surface_lifts"].fillna(0) + df["chair_lifts"].fillna(0) + df["gondola_lifts"].fillna(0)
    if "total_lifts" not in df.columns:
        df["total_lifts"] = df["lifts_sum_calc"]
    df["total_lifts"] = df["total_lifts"].fillna(df["lifts_sum_calc"])
else:
    df["lifts_sum_calc"] = np.nan

if "total_lifts" in df.columns:
    denom = df["total_lifts"].replace(0, np.nan)
    if "surface_lifts" in df.columns:
        df["share_surface_lifts"] = (df["surface_lifts"] / denom).fillna(0)
    else:
        df["share_surface_lifts"] = np.nan
    if "chair_lifts" in df.columns:
        df["share_chair_lifts"] = (df["chair_lifts"] / denom).fillna(0)
    else:
        df["share_chair_lifts"] = np.nan
    if "gondola_lifts" in df.columns:
        df["share_gondola_lifts"] = (df["gondola_lifts"] / denom).fillna(0)
    else:
        df["share_gondola_lifts"] = np.nan

# price: log1p
if "price" in df.columns:
    df["price"] = pd.to_numeric(df["price"], errors="coerce")
    df["log_price"] = np.log1p(df["price"].fillna(df["price"].median(skipna=True)))
else:
    df["log_price"] = np.nan

# lift_capacity log
if "lift_capacity" in df.columns:
    med_cap = df["lift_capacity"].median(skipna=True)
    df["log_lift_capacity"] = np.log1p(df["lift_capacity"].fillna(med_cap))
else:
    df["log_lift_capacity"] = np.nan

display(df[[c for c in ["highest_point","lowest_point","altitude_range","total_slopes","share_beginner","total_lifts","log_price","log_lift_capacity"] if c in df.columns or c in df]].head(20))
```

###  Proxy “Demand Score” + target attendance

Pošto nemamo pravu posjećenost, radimo proxy koji kombinira:

- snijeg (rolling mean + current)

- sezona (winter flag)

- veličina resorta (staze, liftovi, kapacitet)

- cijena (može ići + ili -, ovdje ga tretiramo kao “premium attracts demand”, ali s malom težinom)

```python
df = df.copy()

# safe columns
snow_signal = (
    0.55 * df["snow_imputed"].fillna(0) +
    0.35 * df["snow_3m_mean"].fillna(0) +
    0.10 * df["snow_6m_mean"].fillna(0)
)

winter_boost = 1.0 + 0.15 * df["is_winter_month"].fillna(0)

# resort scale signal
scale = 0
if "total_slopes" in df.columns:
    scale += 0.35 * np.log1p(df["total_slopes"].fillna(df["total_slopes"].median(skipna=True)))
if "total_lifts" in df.columns:
    scale += 0.25 * np.log1p(df["total_lifts"].fillna(df["total_lifts"].median(skipna=True)))
if "lift_capacity" in df.columns:
    scale += 0.40 * df["log_lift_capacity"].fillna(0)

# price small influence
price_term = 0.12 * df["log_price"].fillna(0)

# locality score (opcionalno: ako ima lat/long, možemo penalizirati ekstremne missing snow)
df["locality_score"] = 1.0 - 0.35 * df["snow_missing"].fillna(0)

df["demand_score"] = (snow_signal * winter_boost) + scale + price_term
df["demand_score"] = df["demand_score"] * df["locality_score"]

display(df[["month","snow","snow_imputed","snow_3m_mean","demand_score"]].head(10))
print("Demand score summary:")
display(df["demand_score"].describe())
```

```python
df = df.copy()

# rank-based quantile binning => stable even if many same values
r = df["demand_score"].rank(method="average", pct=True)

df["attendance"] = pd.Series(np.where(r <= 1/3, "low",
                         np.where(r <= 2/3, "medium", "high")), index=df.index)

print(df["attendance"].value_counts(dropna=False))

# quick sanity chart
counts = df["attendance"].value_counts()
plt.figure()
plt.bar(counts.index.astype(str), counts.values)
plt.title("Attendance class distribution")
plt.xlabel("attendance")
plt.ylabel("count")
plt.show()
```

### Final feature table + export features.csv

Ovdje biramo što ide u model (X) + target (y).
Izbacujemo “raw text” kolone koje nisu nužne (ali možeš ih ostaviti za report).

```python
df_out = df.copy()

# preporučeni set featurea (model-friendly)
feature_cols = [
    # keys / meta
    "id","resort","country","continent","month",
    # time
    "month_sin","month_cos","is_winter_month","is_shoulder_month",
    # snow
    "snow","snow_missing","snow_imputed","snow_lag_1","snow_lag_2","snow_3m_mean","snow_3m_std","snow_6m_mean",
    # resort profile
    "highest_point","lowest_point","altitude_range",
    "total_slopes","share_beginner","share_intermediate","share_difficult",
    "total_lifts","lift_capacity","share_surface_lifts","share_chair_lifts","share_gondola_lifts",
    "price","log_price","log_lift_capacity",
    # proxy
    "locality_score","demand_score",
    # target
    "attendance"
]

# keep only those existing
feature_cols_present = [c for c in feature_cols if c in df_out.columns]
df_out = df_out[feature_cols_present].copy()

# final cleanup: ensure not empty
print("Final rows:", len(df_out))
print("Final cols:", df_out.shape[1])
display(df_out.head(3))

PROCESSED_DIR.mkdir(parents=True, exist_ok=True)
df_out.to_csv(OUT_PATH, index=False, encoding="utf-8")
print("Saved:", OUT_PATH.resolve())
```
